## Simple Linear Regression
Till now we have considered real world problems that having classes or labels as their solution space but their are also huge number of problems that consider continous solution space and hence we will tackle these problems here and these problems are simply referred to as regression problems and we will consider various ways to visualize them and hence also build model for them for their solution space.

### Preprocessing the Data
Here we just select the data and make it into a form so that we can train the model and hence use it for predicting new points

```python
## here we use the famous housing dataset for pointing out the main steps in Simple Linear Regression
import pandas as pd
df = pd.read_csv('housing.data.txt',header = None,sep = '\s+')
df.columns = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
```

## EDA
EDA or Exploratory Data Analysis is mainly concerned with the concept of visualizing the relations between the different features of the dataset and hence taking out the most important features which explain the dataset properly and hence can be used for generalization of the model.

```python
## Using the required libraries to importing the dataset
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

## Here we scatter plots of different features taken two at a time so that
## we could understand their respectice correlations and hence use them
cols = ['LSTAT','INDUS','NOX','RM','MEDV']
sns.pairplot(df[cols],height = 2.5)
plt.tight_layout()
plt.show()

## Here we just simply plot the Correlation Coefficient and measure which 
## coefficients have better correlation
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale = 1.5)
hm = sns.heatmap(cm,cbar = True,annot = True,square = True,fmt = '.2f',annot_kws = {'size':15},yticklabels = cols,xticklabels = cols)
plt.show()
```

## Model Training
This is the most important step as in this we would select the required features from out EDA step and use those features to train our model and hence use that model for predicting the function of the provided dataset.

```python
## Here we just train a simple Linear Regression Model by implementing a Simple
## Linear Regression Model with gradient Descent as a class and use that to approximate our function
from Linear_Model import LinearRegressionGD
X = df[['RM']].values
y = df['MEDV'].values
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y[:,np.newaxis]).flatten()
lr = LinearRegressionGD()
lr.fit(X_std,y_std)
## Here we just visualize the Cost of the trained model wrt the iterations 
sns.reset_orig()
plt.plot(range(1,lr.n_iter+1),lr.cost_)
plt.ylabel('SSE')
plt.xlabel('Epoch')
plt.show()
## Hence we can see that the cost is decreasing but doesn't decrease that much further
```

```python
## Here we just plot the regression line or decision boundary of the approximated function 
## by training with the model and plotting the line on the given datapoints
def lin_regplot(X,y,model):
    plt.scatter(X,y,c = 'steelblue',edgecolor = 'white',s = 70)
    plt.plot(X,model.predict(X),color = 'black',lw=2)
    return None

lin_regplot(X_std,y_std,lr)
plt.xlabel('Average number of rooms [RM] (standardized)')
plt.ylabel('Price in $1000s [MEDV] (standardized)')
plt.show()
```

## RANSAC
RANSAC or RANdom SAmple Consensus is an algorithm that trains a model that would not get deviated due to the presence of outliers in the model and it would train the model on the inliers of the training dataset by using a simple algorithm and hence we would be able to get a model which could perform better than the previous model by not learning from the outliers and hence would generalize better on the dataset and have better accuracy.

```python
## Implementing a normal Ransac Regressor model and using it to train our model
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor(LinearRegression(),max_trials = 100,min_samples = 50,loss = 'absolute_loss',residual_threshold = 5.0,random_state=0)
ransac.fit(X,y)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
line_X = np.arange(3,10,1)
line_y_ransac = ransac.predict(line_X[:,np.newaxis])
plt.scatter(X[inlier_mask],y[inlier_mask],
            c ='steelblue',edgecolor = 'white',
            marker = 'o',label = 'Inliers')
plt.scatter(X[outlier_mask],y[outlier_mask],
            c = 'limegreen',edgecolor = 'white',
            marker = 's',label = 'Outliers')
plt.plot(line_X,line_y_ransac,color = 'black',lw = 2)
plt.xlabel('Average number of rooms [RM]')
plt.ylabel('Price is $1000s [MEDV]')
plt.legend(loc = 'upper left')
plt.show()
## Hence we can see that the model lays out a better regression line 
## on the inliers present in the dataset 
```

## Evaluation
Here we just implement out the different methods of our evaluating our regression models and hence use them to access our model.In these problems there is a problem of visualizing our trained model as their are huge number of features present in the dataset due to which we couldn't visualize and hence what we use is to plot the residuals vs the predictions plot so that we could find out what patterns are hidden in that.

By this method we are able to understand if there is still something that are model doesn't generalize well to as we could see that in the points visualized as if the model would have generalized then the only thing that we might notice around the zero line would be random noise present in the dataset but if the model doesn't generalize than we would see a particular pattern in the plot which our model wasn't able to generalize to and hence would point to not a good model.

```python
## preprocessing the dataset
from sklearn.model_selection import train_test_split
X = df.iloc[:,:-1].values
y = df['MEDV'].values
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.3,random_state = 0)
## training a simple model
slr = LinearRegression()
slr.fit(X_train,y_train)
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
## now visualizing the plot to see if the model generalizes or not
plt.scatter(y_train_pred,y_train_pred - y_train,c = 'steelblue',marker = 'o',edgecolor = 'white',label = 'Training data')
plt.scatter(y_test_pred,y_test_pred - y_test,c = 'limegreen',marker = 's',edgecolor = 'white',label = 'Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc = 'upper left')
plt.hlines(y=0,xmin=-10,xmax = 50,color = 'black',lw = 2)
plt.xlim([-10,50])
plt.show()
## Hence we can see that the model does not fully generalize and their is still
## some pattern left to geenralize to
```

```python
## These are some of the important metrics that we use for the Error Calculation
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score             ## It is a score which calculates the model generalization and the more it is close to one more better

print('MSE train: %.3f, test: %.3f ' %(mean_squared_error(y_train,y_train_pred),
                         mean_squared_error(y_test,y_test_pred)))
print('R^2 train: %.3f , test: %.3f ' %(r2_score(y_train,y_train_pred),r2_score(y_test,y_test_pred)))
```

## Polynomial Regression
After discussing Linear Regression, we can now understand the importance of Polynomial Regression and thus we can say that we could train our model on the whole feature set but still not achieve a good accuracy score because of the different ways the function of the algorithm could be available and hence we would need to introduce model which not only have features in linear order but also of higher order which could generalize to the model better and hence we would implement a model with Polynomial Regression which would generalize better to the provided dataset.

But the downside with polynomial regression is that if we increase the degree of the model very much we could overfit to the data very easily and hence would lead to a bad model therefore we should minimize to increasing the degree and should use other methods if we can for training the model.

```python
## Making a dataset and using the required function to build the polynomial regression Algorithm
from sklearn.preprocessing import PolynomialFeatures
X = np.array([258.0,270.0,294.0,320.0,342.0,368.0,396.0,446.0,480.0,586.0])[:,np.newaxis]
y = np.array([236.4,234.4,253.8,298.6,314.2,342.2,360.8,368.0,391.2,390.8])
lr = LinearRegression()
pr = LinearRegression()
quadratic = PolynomialFeatures(degree = 2)
## fitting the quadratic model
X_quad = quadratic.fit_transform(X)
## training the linear model
lr.fit(X,y)
X_fit = np.arange(250,600,10)[:,np.newaxis]
y_lin_fit = lr.predict(X_fit)
## training the quadratic model
pr.fit(X_quad,y)
y_quad_fit = pr.predict(quadratic.fit_transform(X_fit))
## Hence plotting the described models and comparing them
plt.scatter(X,y,label = 'training points')
plt.plot(X_fit,y_lin_fit,label = 'linear fit',linestyle = '--')
plt.plot(X_fit,y_quad_fit,label = 'quadratic fit')
plt.legend(loc = 'upper left')
plt.show()
## hence we can see in this case the quadratic model fits better to the dataset and hence we would also see
## this in many real world problems and would have to train the model on those features as per convenience
y_lin_pred = lr.predict(X)
y_quad_pred = pr.predict(X_quad)
## Checking the SSE and R^2 score of the models also points to that that particular model fits better to the dataset
print('Training MSE linear: %.3f, quadratic: %.3f' %(mean_squared_error(y,y_lin_pred),mean_squared_error(y,y_quad_pred)))
print('Training R^2 linear: %.3f, quadratic: %.3f' %(r2_score(y,y_lin_pred),r2_score(y,y_quad_pred)))
```

```python
## Here we can see clearly that increasing the degree of the feature better approximates our model 
## but it might overfit to the dataset and should not be used that often
X = df[['LSTAT']].values
y = df['MEDV'].values
regr = LinearRegression()
quadratic = PolynomialFeatures(degree = 2)
cubic = PolynomialFeatures(degree = 3)
X_quad = quadratic.fit_transform(X)
X_cubic = cubic.fit_transform(X)
X_fit = np.arange(X.min(),X.max(),1)[:,np.newaxis]

regr = regr.fit(X,y)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y,regr.predict(X))

regr = regr.fit(X_quad,y)
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))
quadratic_r2 = r2_score(y,regr.predict(X_quad))

regr = regr.fit(X_cubic,y)
y_cubic_fit = regr.predict(cubic.fit_transform(X_fit))
cubic_r2 = r2_score(y,regr.predict(X_cubic))

plt.scatter(X,y,label = 'training points',color = 'lightgray')

plt.plot(X_fit,y_lin_fit,label = 'linear (d=1), $R^2 = %.2f' % linear_r2,color = 'blue',lw = 2,linestyle = ':')

plt.plot(X_fit,y_quad_fit,label = 'quadratic (d=2), $R^2 = %.2f' % quadratic_r2,color = 'red',lw = 2,linestyle = '-')

plt.plot(X_fit,y_cubic_fit,label = 'cubic (d=3), $R^2 = %.2f' % cubic_r2,color = 'green',lw = 2,linestyle = '--')

plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in $1000s [MEDV]')
plt.legend(loc = 'upper right')
plt.show()
```

```python
## Here we can see that the transforming the dataset in linear scale is possible
## and these techniques should be used more often as they would not lead to a bad model
X_log = np.log(X)
y_sqrt = np.sqrt(y)

X_fit = np.arange(X_log.min()-1,X_log.max()+1,1)[:,np.newaxis]
regr = regr.fit(X_log,y_sqrt)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y_sqrt,regr.predict(X_log))

plt.scatter(X_log,y_sqrt,
            label = 'training points',
            color = 'lightgray')
plt.plot(X_fit,y_lin_fit,
         label = 'linear (d=1), $R^2 = %.2f$' % linear_r2,color = 'blue',lw = 2)

plt.xlabel('log(% lower status of the population [LSTAT])')
plt.ylabel('$\sqrt{Price \; in \; \$1000s \; [MEDV]}$')
plt.legend(loc = 'lower left')
plt.show()
```

## Non Linearity using Random Forests
Here we would see the use of Random Forests in predicting regression models and they tend to work better than normal methods of linear and polynomial regression as these technoques focus on the global learning of the function and would not be able to learn from all misclassified points but in contrast to this the Random Forests Model tend to train on huge number of small learners which would generalize better to the samll pattern learning in the dataset and hence the random forests would learn better generalization of the model by training on these small learners which we typically use as Decision Trees and these trees train on the dataset on the basis of more the information gain in each node or typically decreasing the variance in each node and hence we would be  able to learn a better model.

```python
## importing, training and visualizing a decision tree model on the given dataset
from sklearn.tree import DecisionTreeRegressor
X = df[['LSTAT']].values
y = df['MEDV'].values
tree = DecisionTreeRegressor(max_depth = 3)
tree.fit(X,y)
sort_idx = X.flatten().argsort()
lin_regplot(X[sort_idx],y[sort_idx],tree)
plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in $1000s [MEDV]')
plt.show()
## Hence we can see that a single decision tree wouldn't generalize that well to
## the dataset and hence would be not that good as we can see from the sharp curve
```

```python
## Here we train a random forest on huge number of smaller decision trees and hence train our model
X = df.iloc[:,:-1].values
y = df['MEDV'].values
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.4,random_state=1)
## training and evaluating the model
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators = 1000,criterion = 'mse',random_state=1,n_jobs=-1)
forest.fit(X_train,y_train)
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)
print('MSE train: %.3f, test: %.3f' %(mean_squared_error(y_train,y_train_pred),mean_squared_error(y_test,y_test_pred)))
print('R^2 train: %.3f, test: %.3f' %(r2_score(y_train,y_train_pred),r2_score(y_test,y_test_pred)))
## Hence we can se that model performs good enough on the dataset and could perform better if we tweak the hyperparameters

## Here we would visualize the models residuals and predictions plot and see
## if we generalized the model better 
plt.scatter(y_train_pred,y_train_pred-y_train,c = 'steelblue',marker = 'o',s=35,alpha =0.9,edgecolor='white',label='Training data')
plt.scatter(y_test_pred,y_test_pred-y_test,c='limegreen',marker='s',s=35,edgecolor = 'white',alpha = 0.9,label = 'Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc = 'upper left')
plt.hlines(y=0,xmin=-10,xmax=50,lw=2,color='black')
plt.xlim([-10,50])
plt.show()
## Hence we can see that  the model still remains to capture some type of pattern in the dataset
```
