## K - Nearest Neighbors
This is the implementation of one of the most widely used classification algorithm and most ppopularly known as the "lazy learner" because of it's nature not to train on the model and generalize but to learn the model dataset and hence achieve the required performance.

This model has many pros and cons with respect to the models we have implemented before and hence it could be used as per the required dataset, We have also shown the required parameter tuning and how to aachieve the better model for the required dataset.

We have also used the required Scikit-Learn Library modules for data preprocessing and training the required model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from Decision_Boundary import plot_decision_boundary
```

### Preprocessing the Data

```python
iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.3,random_state=1,stratify=y)
sc = StandardScaler()
sc.fit(X_train)
X_standard_train = sc.transform(X_train)
X_standard_test = sc.transform(X_test)
X_standard_combined = np.vstack((X_standard_train,X_standard_test))
y_combined = np.hstack((y_train,y_test))
```

### Training the Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5,p=2,metric = 'minkowski')
knn.fit(X_standard_train,y_train)
plot_decision_boundary(X_standard_combined,y_combined,classifier=knn,test_idx=range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc = 'upper left')
plt.show()
```
